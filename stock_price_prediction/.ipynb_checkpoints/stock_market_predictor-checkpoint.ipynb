## Preparing the Dataset

we’ll identify a target that we’re trying to predict. Our target will be if the price will go up or down tomorrow. If the price went up, the target will be 1, and if it went down, the target will be 0.

```python
apple['Target'] = apple.rolling(2).apply(lambda x: x.iloc[1] > x.iloc[0])['Close']
```

## Preparing the Model

```python
X = apple.iloc[1:,[0,1,2,3,4]]
y = apple.iloc[1:, [7]]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Training the Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_random_forest = RandomForestClassifier(n_estimators = 100, min_samples_split = 200)
```

```python
model_random_forest.fit(X_train, y_train)
```

```python
# the parameters on which the model been trained
model_random_forest.get_params()
```

## Predict the Data

```python
rfc_pred = model_random_forest.predict(X_test)
```

```python
rfc_pred
```

```python
# prediction with residuals
for i in range(len(rfc_pred)):
    residuals = y_test - rfc_pred[i]
```

```python
residuals
```

```python
# plot the residuals
sns.displot(residuals, kind = 'kde')
```

#### As we can see we obtain two smooth curves for the two values

## Check the Performance of Our Model

```python
from sklearn.metrics import precision_score
```

```python
score = precision_score(rfc_pred, y_test)
```
